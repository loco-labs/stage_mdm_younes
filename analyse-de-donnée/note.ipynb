# analyse

```python
import pandas as pd
```

```python
team=pd.read_csv("team_activity.csv")
```

```python
analyse=pd.read_csv("workday_context.csv")
analyse
```

```python
df=pd.merge(
    team,
    analyse,
    on="date",
    how="inner"
)
df['date'] = df['date'].astype("datetime64[ns]")
print(df.dtypes)
df["date"].dt.weekday
#df
```

```python
df.groupby("music_playing")["tasks_completed"].mean()
```

```python
df.groupby("meetings_count")["focus_score"].mean()
```

```python
df.groupby(df["date"].dt.strftime("%W"))["tasks_completed"].mean()
```
